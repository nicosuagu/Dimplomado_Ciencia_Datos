```python
import pandas as pd

class MyIterable:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)
        self.i = 0
    
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.i >= len(self.df):
            raise StopIteration
        else:
            result = self.df.iloc[self.i:self.i+2, :]
            self.i += 2
            return result
```

```python
# Crear una instancia de MyIterable y recorrerla
my_iterable = MyIterable('iris.data')
for datos in my_iterable:
    print(datos)
```

```python
import pandas as pd

class MyDataFrame:
    def __init__(self, file_path):
        self.df = pd.read_csv(file_path)

    def basic_statistics(self):
        return self.df.describe()

    def get_column_names(self):
        return list(self.df.columns)

    def get_variable_statistics(self, column_name):
        result = {
            'mean': self.df[column_name].mean(),
            'median': self.df[column_name].median(),
            'std': self.df[column_name].std(),
            'q25': self.df[column_name].quantile(0.25),
            'q50': self.df[column_name].quantile(0.5),
            'q75': self.df[column_name].quantile(0.75)
        }

        self.df[column_name].plot.hist()
        return result
```

```python
# Crear una instancia de MyDataFrame y llamar a sus m√©todos
my_dataframe = MyDataFrame('iris.data')
print(my_dataframe.basic_statistics())
print(my_dataframe.get_column_names())
```
